```python
import pandas as pd
```

```python
def sort_dataframes_by_mean(data, column):
    means = []
    maxes = []
    for df in data: # iterating over df list
        means.append(df[column].mean()) # calculating data
        maxes.append(df[column].max())

    avg_max_df = pd.DataFrame( # creating df with means and maxes
        {'means': means,
         'maxes': maxes}
    )

    avg_max_df.sort_values(by='maxes', inplace = True) # sorting df
    avg_max_df.sort_values(by='means', inplace = True)

    ordered_data = [data[i] for i in avg_max_df.index] # rearranging data by df indexes

    return ordered_data
```
